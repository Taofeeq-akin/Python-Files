````python
## Mimicking a bank

class Bank:
  def __init__(self, account_name, account_number, balance = 0):
    self.name = account_name
    self.balance = balance
    if len(account_number) != 10:
      raise ValueError("Invalid Account Number")
    else:
      self.number = account_number

  def deposit(self, amount):
    self.balance = self.balance + amount
    return f"Y0ur new balance is {self.balance}"

  def withdrawal(self, amount):
    if amount > self.balance:
        raise ValueError("Insufficient Funds")
    else:
      self.balance = self.balance - amount
      return f"Y0ur new balance is {self.balance}"
````

## Transfer from a user to another user in the same bank

````python
## For each Customers
class Customer:
  def __init__(self, account_name, account_number, balance = 0):
    self.account_name = account_name
    self.balance = balance
    if len(account_number) != 10:
      raise ValueError("Invalid Account Number")
    else:
      self.account_number = account_number

  # Self deposit
  def deposit(self, amount):
    self.balance += amount
    return f"[{self.account_name}]: Y0ur new balance is {self.balance}"

  # Self Withdrawal
  def withdrawal(self, amount):
    if amount > self.balance:
        raise ValueError("Insufficient Funds")
    else:
      self.balance -= amount
      return f"[{self.account_name}]: Y0ur new balance is {self.balance}"
````

````python
class Bank:
  def __init__(self):
    # Note this a class attribute and a distinct
    self.customers = {}

  # This for adding new customers
  def add_customer(self, customer):
    if customer in self.customers:
      raise ValueError("Customer already exists")
    else:
       # Customer account number is used for each customer key
      self.customers[customer.account_number] = customer

  def get_customer(self, account_number):
    if account_number in self.customers:
      return self.customers[account_number]
    else:
      raise ValueError("Customer does not exist")

  def transfer(self, account_sender, acct_recipient, amount):
    sender = self.get_customer(account_sender)
    recipient = self.get_customer(acct_recipient)
    sender.withdrawal(amount)
    recipient.deposit(amount)

    return f"{sender.account_name} sent {amount} to {recipient.account_name}"
````

````python
customer1 = Customer("Taofeeq Akin", "2894462753", 10000)
customer2 = Customer("John Doe", "8265389754", 500000)
````
